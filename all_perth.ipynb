```python
import pandas as pd
import numpy as np
import plotly.express as pe
import plotly.express as px
import plotly.graph_objects as pg
import seaborn as sns
#sns.set(style = 'dargrid')
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
data = pd.read_csv('all_perth_310121.csv')
data.head()
```

```python
#which 3 suburbs will you advice a low income earner to relocate to. 
dt=data[['SUBURB','PRICE']]
dt.groupby(['SUBURB'])[['PRICE']].mean().sort_values(by= 'PRICE', ascending=True)
```

```python
#What is the average number of bedrooms in the highbrow areas where prices are above 2 million

dt = data[['SUBURB','BEDROOMS','PRICE']]
v = dt.loc[(data['PRICE'] > 200000)]
v.groupby(['SUBURB','PRICE'])[['BEDROOMS']].mean().sort_values(by= 'BEDROOMS', ascending=False )
```

```python
#What is the average number of bedrooms in the low income areas where price below 150,000
dt = data[['SUBURB','BEDROOMS','PRICE']]
v = dt.loc[(data['PRICE'] < 150000) ]
v.groupby(['SUBURB','PRICE'])[['BEDROOMS']].mean().sort_values(by= 'BEDROOMS', ascending=False )
```

```python
#What is the price of the most expensive house in Westminster Suburb.

dy =data[['SUBURB','BEDROOMS','PRICE']]
x = dy.loc[(data['SUBURB'] == 'Westminster')]
x.groupby(['SUBURB','BEDROOMS'])[['PRICE']].mean().sort_values(by= 'PRICE', ascending=False)
```

```python
#Westminster or Girrawheen Suburb, who has the most expensive average mean houses.
dx = data[['SUBURB', 'PRICE','BEDROOMS']]
v =dx.loc[((data['SUBURB'] == 'Westminister')|(data['SUBURB'] == 'Girrawheen'))]
v.groupby([ 'SUBURB','BEDROOMS'])[['PRICE']].mean().sort_values(by= 'PRICE', ascending=False)
```

```python
#What is the name of the station that is the closest to the biggest land area in the suburb with the lowest price
df =data[['SUBURB','LAND_AREA','NEAREST_STN','PRICE']]
df.groupby(['SUBURB','LAND_AREA','NEAREST_STN'])[['PRICE']].mean().sort_values(by= 'PRICE',ascending=True)
```
